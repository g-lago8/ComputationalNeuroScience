```python
#imports here

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for (A) tonic spiking
a = 0.02
b = 0.2
c = -65
d = 6
V = -70  # Initial membrane potential
u = b * V  # Initial recovery variable
tau = 0.25  # Time step
tspan = np.arange(0, 100 + tau, tau)  # Time span
T1 = tspan[-1] / 10  # Time after which the input current changes

# Initialize lists to store the membrane potential (V) and recovery variable (u) over time
VV = []  # Membrane potential
uu = []  # Recovery variable

# Simulation loop
for t in tspan:
    # Apply input current based on the condition
    I = 14 if t > T1 else 0
    # Update equations
    V = V + tau * (0.04 * V ** 2 + 5 * V + 140 - u + I)
    u = u + tau * a * (b * V - u)
    # Reset condition
    if V > 30:
        VV.append(30)
        V = c
        u = u + d
    else:
        VV.append(V)
    uu.append(u)

# Plotting
plt.figure(figsize=(12, 6))
plt.subplot(1, 1, 1)
plt.plot(tspan, VV)
plt.plot([0, T1, T1, max(tspan)], [-90, -90, -80, -80], 'k--')  # Illustrate the input current change
plt.axis([0, max(tspan), -90, 30])
plt.title('(A) tonic spiking')
plt.axis('off')  # Hide axes
plt.show()
```
